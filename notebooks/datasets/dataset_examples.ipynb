```python
%load_ext autoreload
%autoreload 2

# Local imports
from sv_machines.datasets.circle import get_circle_dataset, get_circle_plot
from sv_machines.datasets.line import get_line_dataset, get_line_plot, add_line_outliers
from sv_machines.datasets.function_a import get_function_a_dataset, get_function_a_plot
```

# Line dataset

```python
slope, offset, epsilon = 2.0, 1.5, 1.5

x_line, y_line = get_line_dataset(
    slope=slope, offset=offset, epsilon=epsilon, num_points=500, epsilon_strict=True
)

fig, ax = get_line_plot(
    x_line, y_line, slope=slope, offset=offset, epsilon=epsilon, with_true_function=True
)
fig.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```

Add outliers to the dataset

```python
x_line_with_outliers, y_line_with_outliers, outliers_indexes = add_line_outliers(
    x_line,
    y_line,
    slope=slope,
    offset=offset,
    epsilon=epsilon,
    num_outliers=30,
    outlier_max_distance=epsilon * 2,
)

fig_outliers, ax_outliers = get_line_plot(
    x_line_with_outliers,
    y_line_with_outliers,
    slope=slope,
    offset=offset,
    epsilon=epsilon,
    with_true_function=True,
    outliers_indexes=outliers_indexes,
)
fig_outliers.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```

# Circle dataset

```python
radius, epsilon = 5.0, 0.5

x_circle, y_circle = get_circle_dataset(
    radius=radius, epsilon=epsilon, num_points=500, epsilon_strict=True
)

fig, ax = get_circle_plot(
    x_circle, y_circle, radius=radius, epsilon=epsilon, with_true_function=True
)
fig.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```
