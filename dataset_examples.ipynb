```python
%load_ext autoreload
%autoreload 2

# Third party imports
import numpy as np
import matplotlib.pyplot as plt

# Local imports
from sv_machines.datasets.circle import get_circle_dataset
from sv_machines.datasets.line import get_line_dataset
```

# Line dataset

```python
slope, offset, epsilon = 2.0, 1.5, 1.5

x_line, y_line = get_line_dataset(slope=slope, offset= offset, epsilon=epsilon, num_points=500, epsilon_strict=True)

x_line_ordered = np.linspace(min(x_line),max(x_line),1000)
y_true = slope*x_line_ordered + offset
y_upper_limit =  slope*x_line_ordered + offset + epsilon
y_lower_limit = slope*x_line_ordered + offset - epsilon

fig, ax = plt.subplots(1,1)
ax.plot(x_line_ordered, y_true, color='red', linestyle='-', label='True y values (without noise)')
# ax.plot(x_line_ordered, y_upper_limit, color='green', linestyle='--', linewidth=2, label='True y values + epsilon')
# ax.plot(x_line_ordered, y_lower_limit, color='green', linestyle='--', linewidth=2, label='True y values - epsilon')
plt.fill_between(x_line_ordered, y_lower_limit, y_upper_limit, color='black', alpha=0.2, label='epsilon band')
ax.scatter(x_line, y_line, marker='x', s=8, color='blue', label='Line dataset points')
fig.tight_layout() 
fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig.show()
```

# Circle dataset

```python
radius, epsilon = 5.0, 0.5

x_circle, y_circle = get_circle_dataset(radius=radius, epsilon=epsilon, num_points=500, epsilon_strict=True)

x_angle_true = np.linspace(0,2*np.pi,100)
x_axis = radius * np.cos(x_angle_true)
y_true = radius * np.sin(x_angle_true)
x_upper_limit = (radius+epsilon) * np.cos(x_angle_true)
y_upper_limit = (radius+epsilon) * np.sin(x_angle_true)
x_lower_limit = (radius-epsilon) * np.cos(x_angle_true)
y_lower_limit = (radius-epsilon) * np.sin(x_angle_true)

fig, ax = plt.subplots(1,1)
ax.plot(x_axis, y_true, color='red', linestyle='-', label='True y values (without noise)')
ax.plot(x_upper_limit, y_upper_limit, color='green', linestyle='--', linewidth=2, label='True y values + epsilon')
ax.plot(x_lower_limit, y_lower_limit, color='green', linestyle='--', linewidth=2, label='True y values - epsilon')
# ax.fill(np.concatenate([x_upper_limit, x_lower_limit[::-1]]),
#         np.concatenate([y_upper_limit, y_lower_limit[::-1]]),
#         color='black', alpha=0.2, label=f'{epsilon} band')
ax.scatter(x_circle, y_circle, marker='x', s=8, color='blue', label='Circle dataset points')
fig.tight_layout() 
fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_aspect('equal') 
fig.show()
```
